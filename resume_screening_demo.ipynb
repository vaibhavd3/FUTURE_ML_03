# Resume Screening and Ranking (Demo)

This notebook builds a small, end-to-end ML-based resume screening system.
It cleans resume text, extracts skills, compares candidates to a job description,
and ranks resumes by role fit.

**Note:** This demo uses built-in sample resumes, no dataset required.

## What this notebook does
- Loads resume text from a CSV (or uses a small demo dataset)
- Cleans and normalizes resume text
- Extracts skills using a keyword-based NLP approach
- Computes resume-to-job similarity scores
- Ranks candidates and highlights missing skills

```python
import re
from typing import List, Dict

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Optional: use spaCy if installed (not required for the demo)
try:
    import spacy
    _NLP = spacy.load("en_core_web_sm")
except Exception:
    _NLP = None

# ---- Configuration ----
USE_DEMO_DATA = True  # Set to False if you want to load a CSV later
DATASET_PATH = r"YOUR_DATASET_PATH.csv"  # Only used when USE_DEMO_DATA = False
TEXT_COLUMN_CANDIDATES = ["Resume", "Resume_str", "resume", "text", "resume_text"]
JOB_DESCRIPTION = (
    "We are hiring a Data Scientist with strong Python, SQL, and ML experience. "
    "Required skills include: pandas, numpy, scikit-learn, statistics,"
    "data visualization, and model deployment. Experience with NLP is a plus."
)
```

```python
def _find_text_column(df: pd.DataFrame) -> str:
    for col in TEXT_COLUMN_CANDIDATES:
        if col in df.columns:
            return col
    # Fallback: take first object column
    for col in df.columns:
        if df[col].dtype == object:
            return col
    raise ValueError("No text column found in dataset.")

def load_resumes(path: str, use_demo: bool = True) -> pd.DataFrame:
    if use_demo:
        demo = [
            "Data Scientist with Python, SQL, pandas, numpy, scikit-learn. Built ML models and dashboards."
            "Software Engineer skilled in Java, Spring, REST APIs, and microservices."
            "Analyst with Excel, SQL, Power BI, statistics, and data visualization experience."
            "NLP Engineer with Python, spaCy, Transformers, and deployment on Docker."
            "ML Engineer with Python, ML pipelines, model deployment, and monitoring."
            "Business Analyst with SQL, Excel, stakeholder communication, and KPI reporting."
            "Data Engineer with Python, SQL, Spark, ETL pipelines, and cloud data warehouses."
            "Applied Scientist with statistics, A/B testing, and ML model deployment in production."
        ]
        df = pd.DataFrame({"resume_text": demo})
        df["candidate_id"] = [f"C{i+1:03d}" for i in range(len(df))]
        return df
    df = pd.read_csv(path)
    text_col = _find_text_column(df)
    df = df[[text_col]].rename(columns={text_col: "resume_text"})
    df["candidate_id"] = [f"C{i+1:03d}" for i in range(len(df))]
    return df

resumes = load_resumes(DATASET_PATH, USE_DEMO_DATA)
resumes.head()
```
